```python
from IPython.display import display
import pandas as pd
import numpy as np
```

# Assumptions

```python
PRob = prob_1.set_index('Years')
display(PRob)
```

```python
#Input the age

while True:
    try:
        age = int(input('Enter age: '))
        if age >= 25 and age <= 50:
            break;
        else:
            print('Invalid entry, age should be between 25-50')
            
    except ValueError:
        print('The age should not be a decimal!')
        continue
        
        
#Input the term

while True:
    try:
        term = int(input('Enter the term you wish: '))
        if (term >= 1 and term <= 25 and term + age <= 50):
            break;
        else:
            print('Invalid!, the maximum you can have is 25')
            
    except ValueError:
        print('The age should not be a decimal!')
        continue
```

```python
b = age + term
```

```python
To_use = PRob.loc[ age : b,:]
display(To_use)
```

# Different assumptions

```python
r = 0.105        # rate of interest
m = 0.03         # management costs
c = 0.02  # commission costs
Con = 244000   # Contributions
S = 1000000
n = 0
```

# Important assumptions

```python
Assumptions = pd.DataFrame(columns = ['years',
                                      'Interest rate',
                                      'Commission charge',
                                      'Mgt charge',
                                      'Sum assured'])
Assumptions['years'] = range(age, b+1) 
Assumptions['Interest rate'] = r
Assumptions['Commission charge'] = c
Assumptions['Mgt charge'] = m
Assumptions['Sum assured'] = S
            
Assumptions = Assumptions.set_index('years')
display(Assumptions)
```

```python
Fund = pd.DataFrame(columns = ['years',
                                'Term',
                                'Value at the start',
                                'contributions',
                                'Expenses',
                                'Interest earned',
                                'Fund after interest',
                                'Extra death cost',
                                'Extra disability cost',
                                'Fund after expenses',
                                'Management charge',
                                'Fund value at the end',
                                'Accumulated value of the company',
                                'NPV of the fund',
                                'NPV of the company',
                                'Profit margin',
                                'Value per contribution '])


Fund['years'] = range(age, b+1)
Fund['Term']  = range(0, term+1)
Fund['contributions'] = Con
Fund['Expenses'] = Fund['contributions'] * c
Fund = Fund.set_index('years')
Fund.at[age,'Value at the start'] = 0
Fund.at[age,'Extra disability cost'] = 0
Fund['Interest earned'] = ((Fund['contributions'])+Fund['Value at the start'])*r
Fund['Fund after interest'] = Fund['Value at the start']+(Fund['contributions']-Fund['Expenses'])+Fund['Interest earned']
Fund['Extra death cost'] = np.where (Fund['Fund after interest'] < Assumptions['Sum assured'], (Assumptions['Sum assured'] - Fund['Fund after interest'])*To_use['(aq)x d'], 0)

Fund['Fund after expenses'] = Fund['Fund after interest']-(Fund['Extra death cost']+Fund['Extra disability cost'])   
Fund['Management charge'] = (Fund['Fund after expenses'])*m
Fund['Fund value at the end'] = Fund['Fund after expenses']-(Fund['Management charge'])

for i in range(age,age+term):
    Fund.loc[i+1,'Value at the start'] = Fund.loc [i,'Fund value at the end']
    
Fund['Interest earned'] = ((Fund['contributions'])+Fund['Value at the start'])*r
    
Fund['Accumulated value of the company'] = Fund['Management charge']*((1+r)**(term - Fund['Term']))
Fund['NPV of the fund'] = Fund['Fund value at the end'] * (1+r)**(-(Fund['Term']+1))
Fund['NPV of the company'] = Fund['Accumulated value of the company'] * (1+r)**(-(Fund['Term']+1))
Fund['Profit margin'] = Fund['NPV of the company']/Fund['NPV of the fund']
Fund['Value per contribution'] = Fund['Fund value at the end']/Assumptions['Sum assured']
```

```python
def my_function():
    for i in range(age,age+term):
         Fund.loc[i+1,'Value at the start'] = Fund.loc [i,'Fund value at the end'] 
    Fund['Interest earned'] = ((Fund['contributions'])+Fund['Value at the start'])*r
    Fund['Fund after interest'] = Fund['Value at the start']+(Fund['contributions']-Fund['Expenses'])+Fund['Interest earned']
    Fund['Extra death cost'] = np.where (Fund['Fund after interest'] < Assumptions['Sum assured'], (Assumptions['Sum assured'] - Fund['Fund after interest'])*To_use['(aq)x d'], 0)
    for i in range(age,b+1):
        if i in range(age,age+3):
            Fund.at[i, 'Extra disability cost'] = 0
        else:
            if Fund.at[i,'Fund after interest'] < Assumptions.at[i,'Sum assured']:
                Fund.at[i,'Extra disability cost'] =  (Assumptions.at[i,'Sum assured'] - Fund.at[i,'Fund after interest'])*To_use.at[i,'(aq)x dis']
            else:
                Fund.at[i,'Extra disability cost'] = 0 
                
    Fund['Fund after expenses'] = Fund['Fund after interest']-(Fund['Extra death cost']+Fund['Extra disability cost'])   
    Fund['Management charge'] = (Fund['Fund after expenses'])*m
    Fund['Fund value at the end'] = Fund['Fund after expenses']-Fund['Management charge']
    Fund['Accumulated value of the company'] = Fund['Management charge']*((1+r)**(term - Fund['Term']))
    Fund['NPV of the fund'] = Fund['Fund value at the end'] * (1+r)**(-(Fund['Term']+1))
    Fund['NPV of the company'] = Fund['Accumulated value of the company'] * (1+r)**(-(Fund['Term']+1))
    Fund['Profit margin'] = Fund['NPV of the company']/Fund['NPV of the fund']
    Fund['Value per contribution'] = Fund['Fund value at the end']/Assumptions['Sum assured']
```

```python
for i in range(age,b+1):
    my_function()
    
    
display(Fund)
```

# Profit Testing

```python
Age_Profit =Fund[['Term','Profit margin']]

dq = Age_Profit.reset_index()
display(dq)
```

```python
import matplotlib.pyplot as plt
dq.plot(x="years", y="Profit margin")

plt.xlabel('Age')
plt.ylabel('Profit Margin')
plt.title('Profit testing')
plt.show()
```
